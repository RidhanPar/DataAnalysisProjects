```python
# Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
# Load the dataset
df = pd.read_csv(r'C:\Users\manvi\Downloads\Dataset\movies.csv')
# Preview the data
df.head()
```

```python
# Print data types
print(df.dtypes)

# Plot boxplot for 'gross' to identify outliers
df.boxplot(column=['gross'])
```

```python
# Remove duplicate rows
df.drop_duplicates()
```

```python
# Sort data by 'gross' in descending order for better analysis
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Regression plot between 'gross' and 'budget'
sns.regplot(x="gross", y="budget", data=df)

# Regression plot between 'score' and 'gross'
sns.regplot(x="score", y="gross", data=df)
```

```python
# Display correlation matrices for different methods
print(df.corr(method='pearson'))
print(df.corr(method='kendall'))
print(df.corr(method='spearman'))

# Heatmap of Pearson correlation matrix
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation matrix for Numeric Features")
plt.xlabel("Movie features")
plt.ylabel("Movie features")
plt.show()
```

```python
# Factorize categorical variables and compute Pearson correlation
correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation matrix for Movies")
plt.xlabel("Movie features")
plt.ylabel("Movie features")
plt.show()
```

```python
# Identify and print strong correlations
correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()
corr_pairs = correlation_mat.unstack()
sorted_pairs = corr_pairs.sort_values(kind="quicksort")
strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]
print(strong_pairs)
```

```python
# Analyze top companies by gross revenue
CompanyGrossSum = df.groupby('company')[["gross"]].sum()
CompanyGrossSumSorted = CompanyGrossSum.sort_values('gross', ascending = False)[:15]
CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 
CompanyGrossSumSorted
```

```python
# Scatter plot for budget vs gross earnings
plt.scatter(x=df['budget'], y=df['gross'], alpha=0.5)
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()
```

```python
# Numerize categorical data and display final correlation matrix
df_numerized = df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category').cat.codes

correlation_matrix = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation matrix for Movies")
plt.xlabel("Movie features")
plt.ylabel("Movie features")
plt.show()
```
